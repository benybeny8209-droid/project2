```python
import numpy as np
import matplotlib.pyplot as plt

# ---   ---
n = np.array([-2, -1, 0, 1])
x = np.array([-1, 1, 1, 2])

#   
plt.stem(n, x)
plt.title("  x[n]")
plt.xlabel("n")
plt.grid(True)
plt.show()

# --- y1[n] = x[n + 3] ---
n1 = n - 3
plt.stem(n1, x)
plt.title("y1[n] = x[n + 3]  (    )")
plt.xlabel("n")
plt.grid(True)
plt.show()

# --- y2[n] = x[n - 1] ---
n2 = n + 1
plt.stem(n2, x)
plt.title("y2[n] = x[n - 1]  (    )")
plt.xlabel("n")
plt.grid(True)
plt.show()

# ---   y[n] = x[-n] ---
n_flip = -n
#     
sorted_indices = np.argsort(n_flip)
plt.stem(n_flip[sorted_indices], x[sorted_indices])
plt.title("y[n] = x[-n]  ( )")
plt.xlabel("n")
plt.grid(True)
plt.show()
```
