```python
import numpy as np
import matplotlib.pyplot as plt

# --- سیگنال اصلی ---
n = np.array([-2, -1, 0, 1])
x = np.array([-1, 1, 1, 2])

# رسم سیگنال اصلی
plt.stem(n, x)
plt.title("سیگنال اصلی x[n]")
plt.xlabel("n")
plt.grid(True)
plt.show()

# --- y1[n] = x[n + 3] ---
n1 = n - 3
plt.stem(n1, x)
plt.title("y1[n] = x[n + 3]  (شیفت به چپ ۳ واحد)")
plt.xlabel("n")
plt.grid(True)
plt.show()

# --- y2[n] = x[n - 1] ---
n2 = n + 1
plt.stem(n2, x)
plt.title("y2[n] = x[n - 1]  (شیفت به راست ۱ واحد)")
plt.xlabel("n")
plt.grid(True)
plt.show()

# --- وارون زمانی y[n] = x[-n] ---
n_flip = -n
# مرتب کردن برای نمایش درست
sorted_indices = np.argsort(n_flip)
plt.stem(n_flip[sorted_indices], x[sorted_indices])
plt.title("y[n] = x[-n]  (وارون زمانی)")
plt.xlabel("n")
plt.grid(True)
plt.show()
```
