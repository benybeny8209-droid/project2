```python
import numpy as np
import matplotlib.pyplot as plt

# ---    ---
# (    )
n = np.array([-3, -2, -1, 0, 1, 2, 3])
x = np.array([0, 1, 2, 3, 2, 1, 0])

#    
def plot_signal(n_vals, x_vals, title):
    plt.stem(n_vals, x_vals)
    plt.title(title)
    plt.xlabel("n  t")
    plt.ylabel("x[]")
    plt.grid(True)
    plt.show()

# -------------------------------
# ) x[n - 2]      
n1 = n + 2
plot_signal(n1, x, "1) x[n - 2]      ")

# -------------------------------
# ) x[(3/2)n + 3]     (3/2)      
#        n    
n2 = np.linspace(-6, 6, 25)
x2 = np.interp((3/2)*n2 + 3, n, x, left=0, right=0)
plot_signal(n2, x2, "2) x[(3/2)n + 3]    ")

# -------------------------------
# ) x[n] + x[2n] - 3    
n3 = np.linspace(-6, 6, 25)
x3_1 = np.interp(n3, n, x, left=0, right=0)
x3_2 = np.interp(2*n3, n, x, left=0, right=0)
x3 = x3_1 + x3_2 - 3
plot_signal(n3, x3, "3) x[n] + x[2n] - 3")

# -------------------------------
# ) x(3t - 1/2)    ( )     / 
t = np.linspace(-2, 2, 200)
x_t = np.maximum(0, 1 - abs(t))  #      
y4 = np.interp(3*t - 0.5, t, x_t, left=0, right=0)
plt.plot(t, y4)
plt.title("4) x(3t - 1/2)    ")
plt.xlabel("t")
plt.grid(True)
plt.show()

# -------------------------------
# ) x((1/2)t + 3)    ( )      
y5 = np.interp((1/2)*t + 3, t, x_t, left=0, right=0)
plt.plot(t, y5)
plt.title("5) x((1/2)t + 3)    ")
plt.xlabel("t")
plt.grid(True)
plt.show()

# -------------------------------
# ) x(t - 2) - t          t   
y6 = np.interp(t - 2, t, x_t, left=0, right=0) - t
plt.plot(t, y6)
plt.title("6) x(t - 2) - t      t")
plt.xlabel("t")
plt.grid(True)
plt.show()
```
