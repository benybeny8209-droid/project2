```python
import numpy as np
import matplotlib.pyplot as plt

# --- تعریف سیگنال پایه ---
# (می‌تونی هر مقدار دلخواه بذاری)
n = np.array([-3, -2, -1, 0, 1, 2, 3])
x = np.array([0, 1, 2, 3, 2, 1, 0])

# تابع کمکی برای رسم
def plot_signal(n_vals, x_vals, title):
    plt.stem(n_vals, x_vals)
    plt.title(title)
    plt.xlabel("n یا t")
    plt.ylabel("x[…]")
    plt.grid(True)
    plt.show()

# -------------------------------
# ۱) x[n - 2] → شیفت به راست ۲ واحد
n1 = n + 2
plot_signal(n1, x, "1) x[n - 2] → شیفت به راست ۲ واحد")

# -------------------------------
# ۲) x[(3/2)n + 3] → فشردگی زمانی × (3/2) و شیفت به چپ ۳ واحد
# برای نمایش تقریباً صحیح در گسسته باید n پیوسته‌تر در نظر بگیریم
n2 = np.linspace(-6, 6, 25)
x2 = np.interp((3/2)*n2 + 3, n, x, left=0, right=0)
plot_signal(n2, x2, "2) x[(3/2)n + 3] → فشرده و شیفت‌خورده")

# -------------------------------
# ۳) x[n] + x[2n] - 3 → جمع و کشیدگی
n3 = np.linspace(-6, 6, 25)
x3_1 = np.interp(n3, n, x, left=0, right=0)
x3_2 = np.interp(2*n3, n, x, left=0, right=0)
x3 = x3_1 + x3_2 - 3
plot_signal(n3, x3, "3) x[n] + x[2n] - 3")

# -------------------------------
# ۴) x(3t - 1/2) → فشردگی زمانی (۳ برابر) و شیفت به راست ۱/۶ واحد
t = np.linspace(-2, 2, 200)
x_t = np.maximum(0, 1 - abs(t))  # یه سیگنال مثلثی فرضی برای نمایش
y4 = np.interp(3*t - 0.5, t, x_t, left=0, right=0)
plt.plot(t, y4)
plt.title("4) x(3t - 1/2) → فشردگی و شیفت")
plt.xlabel("t")
plt.grid(True)
plt.show()

# -------------------------------
# ۵) x((1/2)t + 3) → کشیدگی زمانی (۲ برابر) و شیفت به چپ ۳ واحد
y5 = np.interp((1/2)*t + 3, t, x_t, left=0, right=0)
plt.plot(t, y5)
plt.title("5) x((1/2)t + 3) → کشیدگی و شیفت")
plt.xlabel("t")
plt.grid(True)
plt.show()

# -------------------------------
# ۶) x(t - 2) - t → شیفت به راست ۲ واحد و کم کردن t از مقدار سیگنال
y6 = np.interp(t - 2, t, x_t, left=0, right=0) - t
plt.plot(t, y6)
plt.title("6) x(t - 2) - t → شیفت و ترکیب با t")
plt.xlabel("t")
plt.grid(True)
plt.show()
```
